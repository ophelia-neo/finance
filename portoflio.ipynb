```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from pathlib import Path
import os
from IPython.display import display, HTML
import warnings
warnings.filterwarnings('ignore')

def display_dataframe_to_user(title, df):
    """Display dataframe with a formatted title"""
    print(f"\n{title}")
    print("=" * len(title))
    display(df)
    print("\n")

# Configuration
EXCEL_PATH = '/Users/ophelianeo/Downloads/Dataset for Q3b.xlsx'  
BOND_DURATION = 8.0     # rough effective duration for 10Y - weighted average maturity of bond's cash flow
BOND_CONVEXITY = 60.0   # rough convexity (bp^-2 scaled for decimal) - usually btwn 50 to 70
RF_MONTHLY = 0.0        # set to 0 unless you add a T-bill proxy (decimal monthly)

print("üìä MACRO REGIME ANALYSIS")
print("=" * 50)

# Load data
print("üìÅ Loading data from Excel file...")
if not os.path.exists(EXCEL_PATH):
    print(f"‚ùå Error: Excel file '{EXCEL_PATH}' not found!")
    print("Please ensure the Excel file exists at the specified path.")
    raise FileNotFoundError(f"Excel file not found: {EXCEL_PATH}")

try:
    xls = pd.ExcelFile(EXCEL_PATH)
    print(f"‚úÖ Sheet names found: {xls.sheet_names}")
    
    # Load all sheets
    macro = pd.read_excel(xls, sheet_name='Macro')
    prices = pd.read_excel(xls, sheet_name='Prices')
    yld = pd.read_excel(xls, sheet_name='Yield')
    
    # Process dates
    for df in (macro, prices, yld):
        df['Date'] = pd.to_datetime(df['Date'])
        df.sort_values('Date', inplace=True)
    
    print("‚úÖ Data loaded successfully!")
    
except Exception as e:
    print(f"‚ùå Error loading data: {e}")
    raise

# ---------------------------
# MACRO: Build Regimes
# ---------------------------
print("\nüîç Building macro regimes...")

macro = macro.sort_values('Date').reset_index(drop=True)
macro['GDP YOY'] = macro['GDP YOY'].ffill()

# 6m changes for trend signals
macro['CPI_YoY_6m_chg'] = macro['CPI YOY'] - macro['CPI YOY'].shift(6)
macro['GDP_YoY_6m_chg'] = macro['GDP YOY'] - macro['GDP YOY'].shift(6)

def trend_state(x):
    if pd.isna(x):
        return np.nan
    return 'Rising' if x > 0 else ('Falling' if x < 0 else 'Flat')

macro['InflationTrend'] = macro['CPI_YoY_6m_chg'].apply(trend_state)
macro['GrowthTrend']    = macro['GDP_YoY_6m_chg'].apply(trend_state)
macro['RecessionFlag']  = (macro['GDP YOY'] <= 0).astype(int)

def label_regime(row):
    if row['RecessionFlag'] == 1:
        return 'Recessionary Pressure'
    it = row['InflationTrend']
    gt = row['GrowthTrend']
    if pd.isna(it) or pd.isna(gt):
        return np.nan
    if it == 'Rising' and gt == 'Falling':
        return 'High Inflation / Slowing Growth'
    if it == 'Falling' and gt == 'Rising':
        return 'Disinflationary Expansion'
    if it == 'Falling' and gt == 'Falling':
        return 'Recessionary Pressure'
    if it == 'Rising' and gt == 'Rising':
        return 'Reflation / Overheating'
    return 'Neutral/Transition'

macro['Regime'] = macro.apply(label_regime, axis=1)

# Create regime analysis tables
regimes = macro[['Date','GDP YOY','CPI YOY','InflationTrend','GrowthTrend','RecessionFlag','Regime']].dropna(subset=['Regime']).reset_index(drop=True)

# Display the main datasets
display_dataframe_to_user("üìà MACRO DATA", macro.head(10))
display_dataframe_to_user("üí∞ PRICES DATA", prices.head(10))
display_dataframe_to_user("üìä YIELD DATA", yld.head(10))

# Regime Analysis
print("\nüéØ REGIME ANALYSIS RESULTS")
print("=" * 40)

# Regime counts
regime_counts = regimes['Regime'].value_counts().rename_axis('Regime').reset_index(name='Months')
display_dataframe_to_user("üìä Regime Distribution (Number of Months)", regime_counts)

# Regime date ranges
regime_ranges = regimes.groupby('Regime')['Date'].agg(['min','max']).reset_index()
regime_ranges.columns = ['Regime', 'Start_Date', 'End_Date']
display_dataframe_to_user("üìÖ Regime Date Ranges", regime_ranges)

# Display recent regime data
display_dataframe_to_user("üîç Recent Regime Classifications (Last 20 months)", 
                         regimes[['Date', 'GDP YOY', 'CPI YOY', 'InflationTrend', 'GrowthTrend', 'Regime']].tail(20))

# Visualization of regimes over time
plt.figure(figsize=(15, 10))

# Plot 1: GDP and CPI trends
plt.subplot(3, 1, 1)
plt.plot(regimes['Date'], regimes['GDP YOY'], label='GDP YoY (%)', linewidth=2, color='blue')
plt.plot(regimes['Date'], regimes['CPI YOY'], label='CPI YoY (%)', linewidth=2, color='red')
plt.axhline(y=0, color='black', linestyle='--', alpha=0.5)
plt.title('Economic Indicators Over Time', fontsize=14, fontweight='bold')
plt.ylabel('Percentage (%)')
plt.legend()
plt.grid(True, alpha=0.3)

# Plot 2: Regime timeline
plt.subplot(3, 1, 2)
regime_colors = {
    'Recessionary Pressure': 'red',
    'High Inflation / Slowing Growth': 'orange', 
    'Disinflationary Expansion': 'green',
    'Reflation / Overheating': 'purple',
    'Neutral/Transition': 'gray'
}

# Create regime timeline
for i, regime in enumerate(regimes['Regime'].unique()):
    if pd.notna(regime):
        regime_data = regimes[regimes['Regime'] == regime]
        plt.scatter(regime_data['Date'], [regime]*len(regime_data), 
                   label=regime, color=regime_colors.get(regime, 'black'), 
                   alpha=0.7, s=20)

plt.title('Macro Regime Timeline', fontsize=14, fontweight='bold')
plt.ylabel('Regime')
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left')
plt.grid(True, alpha=0.3)

# Plot 3: Regime frequency
plt.subplot(3, 1, 3)
regime_counts_sorted = regime_counts.sort_values('Months', ascending=False)
bars = plt.bar(range(len(regime_counts_sorted)), regime_counts_sorted['Months'], 
               color=[regime_colors.get(r, 'gray') for r in regime_counts_sorted['Regime']])
plt.xticks(range(len(regime_counts_sorted)), regime_counts_sorted['Regime'], rotation=45, ha='right')
plt.title('Regime Frequency Distribution', fontsize=14, fontweight='bold')
plt.ylabel('Number of Months')
plt.grid(True, alpha=0.3, axis='y')

# Add value labels on bars
for i, bar in enumerate(bars):
    height = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2., height + 0.1,
             f'{int(height)}', ha='center', va='bottom')

plt.tight_layout()
plt.show()

# Summary Statistics
print("\nüìà SUMMARY STATISTICS")
print("=" * 30)

summary_stats = pd.DataFrame({
    'Total Observations': [len(regimes)],
    'Date Range': [f"{regimes['Date'].min().strftime('%Y-%m-%d')} to {regimes['Date'].max().strftime('%Y-%m-%d')}"],
    'Most Common Regime': [regime_counts.iloc[0]['Regime']],
    'Average GDP YoY': [f"{regimes['GDP YOY'].mean():.2f}%"],
    'Average CPI YoY': [f"{regimes['CPI YOY'].mean():.2f}%"],
    'Recession Months': [regimes['RecessionFlag'].sum()]
}).T
summary_stats.columns = ['Value']
display(summary_stats)

print("\n‚úÖ Analysis complete!")
```

/h1/ Regime Framework

We defined four macroeconomic regimes based on the joint behavior of inflation (CPI trend) and GDP growth:

High Inflation / Slowing Growth ‚Üí Rising CPI, falling GDP

Disinflationary Expansion ‚Üí Falling CPI, rising GDP

Recessionary Pressure ‚Üí Falling CPI, falling GDP

Reflation / Overheating ‚Üí Rising CPI, rising GDP

```python

# ---------------------------
# ASSETS: Resample to Month-End & Compute Returns
# ---------------------------    
# Prices: S&P 500, Gold, USD Index (assume price levels ‚Üí pct returns)
# Make sure CELL INDEX: 0 is executed so that 'prices' is defined
prices_m = prices.set_index('Date').resample('M').last()
asset_cols = [c for c in prices_m.columns if c.lower() != 'date']

# Yields: 10Y yield level in %, to month-end
yld_m = yld.set_index('Date').resample('M').last()

# Compute price-based monthly returns
ret_px = prices_m.pct_change()

# Bond TR approximation from yield level
# Convert % to decimal yields
y = (yld_m['US 10YR Bonds'] / 100.0).copy()
dy = y.diff().fillna(0.0)
# Œîy is in decimals; duration/convexity terms work in decimals
# Add simple monthly carry (y/12)
bond_tr = (-BOND_DURATION * dy) + (0.5 * BOND_CONVEXITY * (dy**2)) + (y.shift(1) / 12.0)
bond_tr.name = 'US 10YR TR (approx)'

rets = ret_px.copy()
rets[bond_tr.name] = bond_tr

# Align to regimes (month-end). Regimes are monthly already, but align to end of month
regimes_m = regimes.set_index('Date').reindex(rets.index).copy()

# ---------------------------
# PERFORMANCE BY REGIME
# ---------------------------
print("\nAnalyzing performance by regime...")

def ann_metrics(x, periods_per_year=12):
    mu = x.mean() * periods_per_year
    sd = x.std() * np.sqrt(periods_per_year)
    sharpe = (mu - RF_MONTHLY * periods_per_year) / (sd if sd != 0 else np.nan)
    return pd.Series({'AnnReturn': mu, 'AnnVol': sd, 'Sharpe': sharpe})

by_regime_stats = (
    rets.join(regimes_m['Regime'])
        .dropna(subset=['Regime'])
        .groupby('Regime')
        .apply(lambda df: df.drop(columns=['Regime'], errors='ignore').apply(ann_metrics), include_groups=False)
)

# Save per-regime performance (will be saved to Excel later)
# Flatten MultiIndex for Excel output
by_regime_stats = by_regime_stats.stack(0).unstack(1).sort_index()

# Extract metric tables from by_regime_stats
annret = by_regime_stats['AnnReturn'].unstack(0)
annvol = by_regime_stats['AnnVol'].unstack(0)
sharpe = by_regime_stats['Sharpe'].unstack(0)

# --- Heatmap of Annualized Returns
fig, ax = plt.subplots(figsize=(10, 5))
im = ax.imshow(annret.values, aspect='auto')
ax.set_xticks(range(len(annret.columns)))
ax.set_xticklabels(annret.columns, rotation=45, ha='right')
ax.set_yticks(range(len(annret.index)))
ax.set_yticklabels(annret.index)
ax.set_title('Annualized Returns by Regime and Asset')
fig.colorbar(im, ax=ax, label='Ann. Return')
plt.tight_layout()
plt.show()

# --- Sharpe bar chart (regimes on x-axis; grouped by asset)
fig, ax = plt.subplots(figsize=(11, 5))
x = np.arange(len(sharpe.index))
width = 0.8 / len(sharpe.columns)
for i, col in enumerate(sharpe.columns):
    ax.bar(x + i*width - 0.4 + width/2, sharpe[col].values, width, label=col)
ax.set_xticks(x)
ax.set_xticklabels(sharpe.index, rotation=20, ha='right')
ax.set_title('Sharpe by Regime (per Asset)')
ax.legend(ncol=2)
plt.tight_layout()
plt.show()

# --- Volatility bar chart (AnnVol)
fig, ax = plt.subplots(figsize=(11, 5))
for i, col in enumerate(annvol.columns):
    ax.bar(x + i*width - 0.4 + width/2, annvol[col].values, width, label=col)
ax.set_xticks(x)
ax.set_xticklabels(annvol.index, rotation=20, ha='right')
ax.set_title('Annualized Volatility by Regime (per Asset)')
ax.legend(ncol=2)
plt.tight_layout()
plt.show()
```

```python
# ---------------------------
# CORRELATIONS BY REGIME (fixed)
# ---------------------------
print("\nüîó CORRELATIONS BY REGIME")
print("=" * 30)

# Align regimes to the monthly return index
regimes_s = regimes.set_index('Date')['Regime']
regimes_s = regimes_s.reindex(ret_px.index).ffill()

# Choose the assets you‚Äôre actually using
ASSETS = ['S&P 500', 'Gold', 'USD Index Spot Rate', 'Bonds_TR_est']
ASSETS = [c for c in ASSETS if c in ret_px.columns]

rets = ret_px[ASSETS].copy()

for r in regimes_s.dropna().unique():
    mask = (regimes_s == r)
    sub = rets.loc[mask]
    if len(sub) > 3:
        print(f"\nüìä Correlation Matrix - {r}")
        display(sub.corr().round(3))

# ---------------------------
# DATA-DRIVEN WEIGHTS (already computed above)
# Ensure WEIGHTS_BY_REGIME_COLS = {regime: weights_dict}
# ---------------------------
# If you didn't run the earlier block, uncomment below to construct them:
# dfR = rets.join(regimes_s.rename('Regime'), how='inner').dropna(subset=['Regime'])
# ... compute data_driven_weights ...
# WEIGHTS_BY_REGIME_COLS = {r: w.to_dict() for r, w in data_driven_weights.items()}

print("\nüíº Asset Allocation by Regime (data-driven):")
try:
    weight_df = pd.DataFrame(WEIGHTS_BY_REGIME_COLS).T
    display(weight_df)
except NameError:
    raise NameError("WEIGHTS_BY_REGIME_COLS not found. Build data-driven weights first.")

# ---------------------------
# REGIME-AWARE ALLOCATION & BACKTEST (fixed)
# ---------------------------
print("\nüìä BACKTEST RESULTS")
print("=" * 25)

# Use previous month's regime to avoid look-ahead bias
regime_lag = regimes_s.shift(1)

# Build weight time series on the same index as rets
W = pd.DataFrame(index=rets.index, columns=ASSETS, dtype=float)

for dt in W.index:
    r = regime_lag.loc[dt]
    wmap = WEIGHTS_BY_REGIME_COLS.get(r, None)
    if wmap is None:
        W.loc[dt] = np.nan
    else:
        # fill weights for available assets; 0 for any missing
        row = {a: float(wmap.get(a, 0.0)) for a in ASSETS}
        s = sum(row.values())
        if s > 0:
            row = {k: v/s for k, v in row.items()}  # renormalize just in case
        W.loc[dt] = pd.Series(row)

W = W.ffill()

# Portfolio returns
port_ret = (W * rets).sum(axis=1)
port_ret.name = 'Regime-Aware'

# 60/40 benchmark (S&P 500 / Bonds_TR_est if available; else fallback)
if 'S&P 500' in rets.columns:
    bond_candidates = [c for c in rets.columns if ('bond' in c.lower()) or ('bonds_tr' in c.lower()) or ('agg' in c.lower())]
    if bond_candidates:
        bond_col = bond_candidates[0]
        bench_ret = 0.60 * rets['S&P 500'].fillna(0) + 0.40 * rets[bond_col].fillna(0)
    else:
        # fallback if no bond proxy available
        bench_ret = rets.mean(axis=1)
else:
    bench_ret = rets.mean(axis=1)

bench_ret.name = '60/40'

# Equity curves
eq = pd.concat([(1+port_ret.fillna(0)).cumprod(),
                (1+bench_ret.fillna(0)).cumprod()], axis=1)
eq.columns = ['Regime-Aware', '60/40']

# Performance summary function
def perf_summary(r):
    r = r.dropna()
    if len(r) == 0:
        return pd.Series({'CAGR': np.nan, 'AnnVol': np.nan, 'Sharpe': np.nan, 'MaxDD': np.nan})
    cagr = (1 + r).prod()**(12/len(r)) - 1
    vol  = r.std() * np.sqrt(12)
    eqc  = (1 + r).cumprod()
    mdd  = (eqc/eqc.cummax() - 1).min()
    rf_m = 0.0 if 'RF_MONTHLY' not in globals() else RF_MONTHLY
    sharpe = (r.mean()*12 - rf_m*12) / (r.std()*np.sqrt(12)) if r.std() != 0 else np.nan
    return pd.Series({'CAGR': cagr, 'AnnVol': vol, 'Sharpe': sharpe, 'MaxDD': mdd})

summary = pd.concat([perf_summary(port_ret), perf_summary(bench_ret)], axis=1)
summary.columns = ['Regime-Aware', '60/40']
display(summary.round(4))

# Equity curve visualization with regime bands (using lagged regime used for trading)
def plot_with_regime_bands(dates, series_dict, regimes_series, title):
    fig, ax = plt.subplots(figsize=(15, 8))
    regime_colors = {
        'Recessionary Pressure': 'lightcoral',
        'High Inflation / Slowing Growth': 'lightsalmon',
        'Disinflationary Expansion': 'lightgreen',
        'Reflation / Overheating': 'plum',
        'Neutral/Transition': 'lightgray'
    }
    # shaded bands
    current = None
    start = None
    labeled = set()
    rs = regimes_series.reindex(dates).ffill()
    for dt, reg in rs.items():
        if reg != current:
            if current is not None and current == current:  # not NaN
                ax.axvspan(start, dt, alpha=0.2,
                           color=regime_colors.get(current, 'lightgray'),
                           label=None if current in labeled else current)
                labeled.add(current)
            current = reg
            start = dt
    if start is not None and isinstance(current, str):
        ax.axvspan(start, dates[-1], alpha=0.2,
                   color=regime_colors.get(current, 'lightgray'),
                   label=None if current in labeled else current)

    for name, s in series_dict.items():
        ax.plot(dates, s.values, label=name, linewidth=2)

    ax.set_title(title, fontsize=16, fontweight='bold')
    ax.set_xlabel('Date'); ax.set_ylabel('Growth of $1')
    ax.legend(bbox_to_anchor=(1.05, 1), loc='upper left')
    ax.grid(True, alpha=0.3)
    plt.tight_layout()
    plt.show()

plot_with_regime_bands(
    eq.index,
    {'Regime-Aware': eq['Regime-Aware'], '60/40': eq['60/40']},
    regime_lag,  # lagged regimes used to trade
    'Equity Curves with Regime Bands (Regime-Aware vs 60/40)'
)

# Display recent values
print("\nüìà Recent Equity Curve Values")
print("=" * 30)
display(eq.tail(12))
```
